```python
top_sell = Data.sort_values(by='Price', ascending=False)
figure = plt.figure(figsize=(20,6))
sns.barplot(y=top_sell.Price, x=top_sell.Price)
plt.xticks()
plt.xlabel('selling_price')
plt.ylabel('seller_type')
plt.title('The selling price of car')
plt.show()
```

```python
def recommend_mouse(x):
    y = Data[["Model","Size","Design","Hand","Price","Weight","Wire","Grip"]][Data["Design" "Size" "Wire" and "Grip"] == x]
    y = y.sort_values(by="Price",ascending=False)
    return y.head(1)
```

```python
recommend_mouse("Ambidextrous" "XS" and "Fingertip")
```

```python
def recommend_mousee(x):
    y = Data[["Design",'Model',"Size","Price"]][Data["Design" and "Size"] == x]
    y = y.sort_values(by="Price",ascending=False)
    return y.head(1)
```

```python
recommend_mousee("Ergonomic" and "L")
```
